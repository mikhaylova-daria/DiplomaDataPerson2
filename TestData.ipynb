```python
dect_type = dict()
for folder in os.listdir(pathCorpus):
    f=open(unicode("/home/daria/Diploma/DiplomaDataPerson/DiplomaDataPerson2/WikiSubset/TestCorpus"+os.sep + folder+os.sep+'features'))
    X = 'id,Word,Offset,TypeNE\n'
    for i, line in enumerate(f):
        line = line.split()
        if len(line)>2:
            word =line[1][len('Text='):]
            if word.find(',')!= -1:
                word = '"'+word+'"'
            line2=str(i)+','
            line2+=(word+','+line[2][len('Offset='):]+',')
            type_w = 'No'
            if (line[0].split(',')[0]=="Person"):
                line2+='Person\n'
                type_w = 'Person'
            else:
                line2+='No\n'
            dect_type[word] = type_w
            X+=line2
    f.close()
    f=open(unicode(pathCorpus+os.sep+folder+os.sep+'features'), 'w')
    f.write(X)
    f.close()
```

```python
import nltk.stem.porter

from time import time

import argparse
import sys

import re
def get_shape(x):
    x = re.sub(r'([a-z]+)','x',x)
    x = re.sub(r'([A-Z]+)','X',x)
    x = re.sub(r'([0-9]+)','0',x)
    return x
```

```python
from collections import OrderedDict
stemmer = nltk.stem.porter.PorterStemmer()
allfolders = os.listdir(pathCorpus)
```

```python
def generarteCSV(folder):
    article = open(unicode(pathCorpus+os.sep+folder+os.sep+"article"))
    text = article.read().decode('utf-16').encode('utf-8')
    dfForArticle = pandas.DataFrame()
    words = []
    offset = []
    pos_in_sent = []
    stem = []
    last_two = []
    last_three = []
    isupper = []
    istitle = []
    isdigit = []
    shape = []
    postager = []
    entities_type=[]
    l = 0
    sentences = nltk.tokenize.sent_tokenize(text.decode('utf-8'))  
    for i, sentence in enumerate(sentences):
        sentence_list = nltk.tokenize.word_tokenize(sentence)
        #pos_tag_list =  nltk.pos_tag(sentence_list)
        for pos_i, word in enumerate(sentence_list):
            word = word.encode('utf-8')
            try:
                entities_type.append(dect_type[word])
            except Exception:
                entities_type.append('No')
            #print word
            try:
                words.append(word)
            except Exception:
                words.append(word)
            pos = text.find(word, l)
            offset.append(pos)
            l = max(len(word) + pos, l)
            if (pos_i<3):
                pos_in_sent.append(pos_i)
            else:
                pos_in_sent.append(-1)
            try:
                stem.append(stemmer.stem(word)),
            except Exception:
                stem.append(word)
            last_two.append(word[-3:]),
            last_three.append(word[-2:]),
            isupper.append(word.isupper()),
            istitle.append(word.istitle()),
            isdigit.append(word.isdigit()),
            shape.append(get_shape(word)),
            #postager.append(pos_tag_list[pos_i][1])
    dfForArticle.insert(0, 'Word', words)
    dfForArticle.insert(1, 'Offset', offset)
    dfForArticle.insert(2, 'Pos_in_sent', pos_in_sent)
    dfForArticle.insert(3, 'Stem', stem)
    dfForArticle.insert(4, 'LastTwo', last_two)
    dfForArticle.insert(5, 'LastThree', last_three)
    dfForArticle.insert(6, 'IsUpper', isupper)
    dfForArticle.insert(7, 'IsTitle', istitle)
    dfForArticle.insert(8, 'IsDigit', isdigit)
    dfForArticle.insert(9, 'Shape', shape)
    dfForArticle.insert(2, 'TypeNE', entities_type)
    dfForArticle.to_csv(path_or_buf=pathCorpus+os.sep+folder+os.sep+"features", index_label="id")
```
